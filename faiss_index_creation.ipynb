```python
from transformers import pipeline
from sentence_transformers import SentenceTransformer, util
import pandas as pd
import faiss
```

```python
# small-scale model for simple embeddings :P
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
path = r"\preset-generator\Themes Dataset\lightroom_theme_configs.csv"
df = pd.read_csv(path)
df
```

```python
themes = df["theme"].tolist()
themes
```

```python
embeddings = model.encode(themes, convert_to_numpy=True)
dimensions = embeddings.shape[1]
index = faiss.IndexFlatL2(dimensions)
index.add(embeddings)

faiss.write_index(index, r"\full_index_faiss.index")
print(f"✅ Generated {index.ntotal} embeddings for {len(themes)} themes.")
```
